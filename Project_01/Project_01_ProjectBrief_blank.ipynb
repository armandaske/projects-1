__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

## Find your next Movie

3. __Filter__ the Dataset for movies that meet the following conditions:

__Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)__

```python
new_df['cast']=df.cast.copy()
bruce_willis_movies=new_df[new_df.cast.str.contains('Bruce Willis').fillna(False)]
bruce_willis_scifi=bruce_willis_movies[bruce_willis_movies.genres.apply(lambda x: ('science fiction' in x.lower()) & ('action' in x.lower()) if isinstance(x,str) else False)].sort_values('vote_average', ascending=False).drop('cast',axis=1)
HTML(bruce_willis_scifi.to_html(escape=False))
```

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

```python
uma_thurman_movies=new_df[(new_df.cast.str.contains('uma thurman', case=False))&(new_df.director.str.contains('quentin tarantino', case=False))].sort_values('runtime').drop('cast',axis=1).head(5)
HTML(uma_thurman_movies.to_html(escape=False))
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
new_df['production_companies']=df['production_companies'].copy()
pixar_movies=new_df[(new_df.production_companies.str.contains('pixar',case=False)) & (new_df.release_date>='2010')  & (new_df.release_date<'2016')].drop('cast',axis=1)
pixar_movies=pixar_movies.sort_values('revenue_musd', ascending=False).head(7)
HTML(pixar_movies.to_html(escape=False))
```

__Search 4: Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent movies first)__

```python
new_df['original_language']=df.original_language.copy()
language_mask=new_df.original_language.str.contains('en', case=False)
rating_mask = new_df.vote_average>=7.5
votes_mask = new_df.vote_count>=10
genre_mask = new_df.genres.str.contains('thriller', case=False) | new_df.genres.str.contains('action', case=False)
subset=new_df[language_mask & rating_mask & genre_mask & votes_mask].drop('cast',axis=1).sort_values('release_date', ascending=False).head(10)
HTML(subset.to_html(escape=False))
```

## Are Franchises more successful?

4. __Analyze__ the Dataset and __find out whether Franchises (Movies that belong to a collection) are more successful than stand-alone movies__ in terms of:

- mean revenue
- median Return on Investment
- mean budget raised
- mean popularity
- mean rating

hint: use groupby()

__Franchise vs. Stand-alone: Average Revenue__

```python
new_df=df.copy()
new_df['is_franchise']=new_df['belongs_to_collection'].notna()
new_df.groupby('is_franchise').revenue_musd.mean()
```

__Franchise vs. Stand-alone: Return on Investment / Profitability (median)__

```python
new_df['ROI']=new_df['revenue_musd']/new_df['budget_musd']
new_df.groupby('is_franchise').ROI.median()
```

__Franchise vs. Stand-alone: Average Budget__

```python
new_df.groupby('is_franchise').budget_musd.mean()
```

__Franchise vs. Stand-alone: Average Popularity__

```python
new_df.groupby('is_franchise').popularity.mean()
```

__Franchise vs. Stand-alone: Average Rating__

```python
new_df.groupby('is_franchise').vote_average.mean()
```

```python
new_df.groupby("is_franchise").agg({"budget_musd": "mean", "revenue_musd": "mean", "vote_average": "mean",
                            "popularity": "mean", "ROI":"median", "vote_count":"mean"})
```

## Most Successful Directors

6. __Find__ the __most successful Directors__ in terms of

- __total number of movies__
- __total revenue__
- __mean rating__

```python
directors=new_df.groupby('director').agg({'title':'count', 'revenue_musd':'sum','vote_average':'mean', 'vote_count':'sum'}).sort_values('revenue_musd',ascending=False)
directors
```

```python
plt.figure(figsize = (12, 8))
directors.nlargest(20,'title').title.plot(kind='bar', fontsize=15)
plt.title("Most Active Directors",fontsize = 20)
plt.ylabel("Number of Movies", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
directors.revenue_musd.nlargest(20).plot(kind='bar', fontsize = 15)
plt.title("Total Revenue",fontsize = 20)
plt.ylabel("Revenue (in MUSD)", fontsize = 15)
plt.show()
```

```python
directors[(directors.title>=5) &  (directors.vote_count>=1000)].nlargest(50,'vote_average')
```

## Release dates by directors named Ford

```python
dates_in_list=new_df.groupby('director')['release_date'].agg(list).to_frame()
dates_expanded=pd.DataFrame(dates_in_list.release_date.to_list())
dates_by_directors = pd.concat([dates_expanded, dates_in_list.reset_index()], axis=1).set_index('director')
# Drop the original 'Names' column
dates_by_directors.drop('release_date', axis=1, inplace=True)
dates_by_directors=dates_by_directors.stack().reset_index(level=0).set_index('director')
dates_by_directors.columns=['release_date']
dates_by_directors[dates_by_directors.index.str.contains(' ford$', case=False)].sort_values('release_date')
```
