## Dropping irrelevant Columns

2. __Drop__ the irrelevant columns 'adult', 'imdb_id', 'original_title', 'video' and 'homepage'.

```python
df_dropped=df_raw.drop(['adult','imdb_id', 'original_title', 'video','homepage'], axis=1)
```

## How to handle stringified JSON columns

3. __Evaluate__ Python Expressions in the stringified columns ["belongs_to_collection", "genres", "production_countries", "production_companies", "spoken_languages"] and __remove quotes__ ("") where possible.

```python
df_evaluated=df_dropped.copy()
df_evaluated['belongs_to_collection']=df_evaluated['belongs_to_collection'].apply(lambda x: eval(x) if isinstance(x,str) else x)
df_evaluated['genres']=df_evaluated['genres'].apply(lambda x: eval(x) if isinstance(x,str) else x)
df_evaluated['production_countries']=df_evaluated['production_countries'].apply(lambda x: eval(x) if isinstance(x,str) else x)
df_evaluated['production_companies']=df_evaluated['production_companies'].apply(lambda x: eval(x) if isinstance(x,str) else x)
df_evaluated['spoken_languages']=df_evaluated['spoken_languages'].apply(lambda x: eval(x) if isinstance(x,str) else x)
df_evaluated
```

## How to flatten nested Columns

4. __Extract__ only the __collection name__ from the column "belongs_to_collection" and __overwrite__ "belongs_to_collection". <br> For example: The value in the first row (Toy Story) should be 'Toy Story Collection'.

```python
df_flattened=df_evaluated.copy()
df_flattened['belongs_to_collection']=df_evaluated['belongs_to_collection'].apply(lambda x: x.get('name') if isinstance(x,dict) else x)
df_flattened
```

5. __Extract__ all __genre names__ from the column "genres" and __overwrite__ "genres". If a movie has more than one genre, __seperate genres by a pipe__ "|".<br>
For example: The value in the first row (Toy Story) should be 'Animation|Comedy|Family'.

```python
def get_items_from_dicts(ls: list, key)->str:
    item_str=''
    for dict_ in ls:
        item_str+=dict_.get(key,'')+'|'
    return item_str.strip('|')
    
df_flattened['genres']=df_evaluated['genres'].apply(lambda x: get_items_from_dicts(x, 'name') if isinstance(x,list) else x)
df_flattened
```

6. __Extract__ all __spoken language names__ from the column "spoken_languages" and __overwrite__ "spoken_languages". If a movie has more than one spoken language, __seperate spoken languages by a pipe__ "|".<br>
For example: The value in the first row (Toy Story) should be 'English'.

```python
df_flattened['spoken_languages']=df_evaluated['spoken_languages'].apply(lambda x: get_items_from_dicts(x, 'name') if isinstance(x,list) else x)
df_flattened
```

7. __Extract__ all __production countries names__ from the column "production_countries" and __overwrite__ "production_countries". If a movie has more than one production country, __seperate production countries by a pipe__ "|".<br>
For example: The value in the first row (Toy Story) should be 'United States of America'.

```python
df_flattened['production_countries']=df_evaluated['production_countries'].apply(lambda x: get_items_from_dicts(x, 'name') if isinstance(x,list) else x)
df_flattened
```

8. __Extract__ all __production companies names__ from the column "production_companies" and __overwrite__ "production_companies". If a movie has more than one production company, __seperate production companies by a pipe__ "|".<br>
For example: The value in the first row (Toy Story) should be 'Pixar Animation Studios'

```python
df_flattened['production_companies']=df_evaluated['production_companies'].apply(lambda x: get_items_from_dicts(x, 'name') if isinstance(x,list) else x)
df_flattened
```

9. __Inspect__ all columns above with value_counts(). Do you see anything strange? __Take reasonable measures__!

## Cleaning Text / String Columns

15. __Analyze__ the text columns "overview" and "tagline". Try to identify __missing data that is not represented by NaN__ (e.g. "No Data"). __Replace as NaN__ (np.nan)!

### cleaning overview

```python
#check values of 3 words or less:
df_valid_overviews=df_flattened.copy()
df_overviews_str=df_valid_overviews[df_valid_overviews.overview.apply(lambda x: True if isinstance(x,str) else False)]
df_overviews_less_3=df_overviews_str[df_overviews_str.overview.apply(lambda x: True if len(x.split())<=3  else False)]
df_overviews_less_3.overview.value_counts().index
```

```python
no_overview_pos=[0,1,2,4,8,10,11,17,19,22,25,28,33]
incorrect_overviews=df_overviews_less_3.overview.value_counts().index[no_overview_pos].to_list()
incorrect_overviews
```

```python
ids_to_modify=df_overviews_str[df_overviews_str.overview.isin(incorrect_overviews)].index
ids_to_modify
```

```python
df_valid_overviews.loc[ids_to_modify,'overview']=np.nan
df_valid_overviews
```

### cleaning tagline

```python
#Valid taglines are naturally short so it will need different tactics. 
#First we will check for very short ones to capture invalid taglines
df_valid_taglines=df_valid_overviews.copy()
df_taglines_str=df_valid_taglines[df_valid_taglines.tagline.apply(lambda x: True if isinstance(x,str) else False)]
df_taglines_less_3=df_taglines_str[(df_taglines_str.tagline.str.split().apply(len)<=2) & (df_taglines_str.tagline.str.len()<=15)]
df_taglines_less_3.tagline.unique()
# '   ' and '-' are the only invalid ones
incorrect_taglines=['-','   ']
```

```python
#Now lets check for taglines that have the word 'no' to catch possible phrases like 'no data', etc
df_taglines_contains_no=df_taglines_str[(df_taglines_str.tagline.str.contains('No ', case=False)) & (df_taglines_str.tagline.str.split().apply(len)<=4)]
df_taglines_contains_no.tagline.unique()
#There are no taglines with 'no data' type of phrases
```

```python
ids_to_modify=df_taglines_less_3[df_taglines_less_3.tagline.isin(incorrect_taglines)].index
df_valid_taglines.loc[ids_to_modify,'tagline']=np.nan
df_valid_taglines
```
